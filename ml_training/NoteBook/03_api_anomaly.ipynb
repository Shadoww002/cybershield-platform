# API Anomaly Detection

Training Isolation Forest for API security.

```python
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import joblib

# Generate synthetic API data
np.random.seed(44)
X = np.random.randn(6000, 56)

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train Isolation Forest
model = IsolationForest(n_estimators=100, contamination=0.1)
model.fit(X_scaled)

# Test
predictions = model.predict(X_scaled)
anomalies = sum(predictions == -1)
print(f'Anomalies detected: {anomalies}/{len(X)}')

# Save
joblib.dump(model, '../../backend/models/api_anomaly_detector.pkl')
joblib.dump(scaler, '../../backend/models/scaler_api.pkl')
print('✓ Model saved!')
```
